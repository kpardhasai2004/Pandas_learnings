#Maps

```python
home_reviews_age_mean = home_reviews.housing_median_age.mean()
home_reviews.housing_median_age.map( lambda p : p - home_reviews_age_mean )
```

```python
def remean_ages(row):
  row.housing_median_age = row.housing_median_age - home_reviews_age_mean
  return row

home_reviews.apply(remean_ages, axis='columns')
```

#Combining

```python
california_housing_test = pd.read_csv("/content/sample_data/california_housing_test.csv")
california_housing_train = pd.read_csv("/content/sample_data/california_housing_train.csv")

pd.concat([ california_housing_test, california_housing_train ])
```

```python
left = california_housing_test.set_index(['longitude','latitude'])
right = california_housing_train.set_index(['longitude','latitude'])

left.join(right, lsuffix='_test', rsuffix='_train')
```
